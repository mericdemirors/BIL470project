```python
import pandas as pd
imdb_data = pd.read_csv("data.csv", sep="\t")
imdb_data.columns = [["tconst", "average point", "total vote number"]]
imdb_data["rating link"] = "https://www.imdb.com/title/"+imdb_data["tconst"]+"/ratings/?ref_=tt_ov_rt"
imdb_data.head(5)
```

```python
def get_votes(link):
    import requests
    text = requests.get(link).text
    s = text[(text.index(("<table cellpadding=\"0\">"))):]
    s = s[0:s.index("</table>")]
    info_text = s.split("\"leftAligned\">")
    votes = {}
    for i,vote in enumerate(info_text[2:]):
        votes[10-i] = vote.split("<")[0]
    return votes
```

```python
def get_first_i_movies_votes(i):
    global imdb_data
    imdb_data["rating votes"] = None
    for i in range(i):
        votes = get_votes(imdb_data.iloc[i]["rating link"])
        imdb_data.iat[i,4] = votes
```
